Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data collection and preprocessing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading data to a dataframe
data_frame = pd.DataFrame(data = breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# separating the features and target
X = data_frame.drop(columns = ['label'], axis = 1)
Y = data_frame['label']
```

```python
print(X.shape, Y.shape)
```

```python
print(Y)
```

```python
# splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
# preprocessing the X_train and X_test
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_test_std)
```

Building the Neural Network

image.png

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# comppiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
plt.show()
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

**Building the predective system**

```python
def cancer_label_prediction(data):

  # change the input_data to a numpy array
  input_data_as_array = np.asarray(data)

  # rehape the numpy array as we are predicting for one data point
  input_data_reshaped = input_data_as_array.reshape(1, -1)

  # standerdizing the input_data
  input_data_std = scaler.transform(input_data_reshaped)

  prediction = model.predict(input_data_std)

  prediction_label = [np.argmax(prediction)]

  return prediction_label
```
